```python
# =========================
# ◊ò◊ï◊§◊° ◊°◊ò◊ï◊ì◊†◊ò◊ô◊ù ◊û◊¢◊ï◊¶◊ë ◊ë-Colab
# =========================

# 1) Mount Google Drive
from google.colab import drive
drive.mount('/content/drive', force_remount=True)

import json, textwrap, os
from pathlib import Path
import ipywidgets as w
from IPython.display import display, HTML

# 2) ◊†◊™◊ô◊ë ◊ú◊ß◊ï◊ë◊• ◊î-JSON
FILE_PATH = Path("/content/drive/MyDrive/students.json")

# 3) ◊ê◊ù ◊ê◊ô◊ü ◊ß◊ï◊ë◊• - ◊†◊ô◊¶◊ï◊® ◊ì◊ï◊í◊û◊î
if not FILE_PATH.exists():
    example = {
        "students": [
            {
                "first_name": "Bahaa",
                "last_name": "Birany",
                "email": "bahaa.birany@e.braude.ac.il",
                "courses": [
                    "Software Engineering Project",
                    "Network Data Analysis",
                    "Partial Differential Equations"
                ],
                "link": "https://www.braude.ac.il/"
            },
            {
                "first_name": "Elias",
                "last_name": "Essawi",
                "email": "elias.essawi@e.braude.ac.il",
                "courses": ["Database Systems", "Artificial Intelligence", "Web Development"],
                "link": "https://www.w3schools.com/"
            },
            {
                "first_name": "Bilal",
                "last_name": "Gadban",
                "email": "bilal.gadban@braude.ac.il",
                "courses": ["Data Security and Cryptology", "Operating Systems", "Advanced Java"],
                "link": "https://www.geeksforgeeks.org/"
            }
        ]
    }
    with open(FILE_PATH, "w", encoding="utf-8") as f:
        json.dump(example, f, ensure_ascii=False, indent=4)

# 4) ◊ß◊®◊ô◊ê◊™ ◊î◊†◊™◊ï◊†◊ô◊ù
with open(FILE_PATH, "r", encoding="utf-8") as f:
    data = json.load(f)

students = data.get("students", [])

# ◊§◊ï◊†◊ß◊¶◊ô◊ô◊™ ◊¢◊ñ◊®: ◊©◊ù ◊û◊ú◊ê
def full_name(s):
    return f"{s.get('first_name','')} {s.get('last_name','')}"

# 5) CSS ◊ß◊ú◊ô◊ú ◊ú◊¢◊ô◊¶◊ï◊ë
style_html = """
<style>
:root { --bg:#0f172a; --card:#111827; --muted:#9ca3af; --accent:#38bdf8; --chip:#1f2937; }
.student-card {
  font-family: system-ui,-apple-system,Segoe UI,Roboto,Helvetica,Arial;
  background: linear-gradient(180deg,#0b1220,#0b1220 40%,#0b1220);
  color: #e5e7eb; border: 1px solid #1f2937; border-radius: 16px;
  padding: 18px; box-shadow: 0 8px 24px rgba(0,0,0,.25); max-width: 820px;
}
.header { display:flex; align-items:center; gap:14px; margin-bottom:8px; }
.avatar {
  width:44px; height:44px; border-radius: 12px; background: #111827; display:flex; align-items:center;
  justify-content:center; font-weight:700; letter-spacing:.5px; border:1px solid #1f2937;
}
.name { font-size:20px; font-weight:700; }
.email a { color:#93c5fd; text-decoration:none; }
.email a:hover { text-decoration:underline; }
.row { display:flex; flex-wrap:wrap; gap:18px; }
.block { background:#0b1220; border:1px solid #1f2937; border-radius:12px; padding:12px 14px; flex:1; min-width:240px; }
.block h4 { margin:0 0 8px 0; font-size:13px; color:#9ca3af; font-weight:600; letter-spacing:.4px; text-transform:uppercase; }
.chips { display:flex; gap:8px; flex-wrap:wrap; }
.chip { background:#111827; border:1px solid #1f2937; border-radius:999px; padding:6px 10px; font-size:13px; color:#e5e7eb; }
.link a { color:#86efac; text-decoration:none; font-weight:600; }
.link a:hover { text-decoration:underline; }
.help { color:#9ca3af; font-size:12px; margin-top:8px; }
.search-wrap { display:flex; gap:10px; align-items:center; }
</style>
"""

# 6) ◊ï◊ï◊ô◊ì◊í'◊ò◊ô◊ù: ◊ó◊ô◊§◊ï◊© + ◊™◊ô◊ë◊™ ◊ë◊ó◊ô◊®◊î + ◊ê◊ñ◊ï◊® ◊™◊¶◊ï◊í◊î
search_box = w.Text(placeholder='◊ó◊ô◊§◊ï◊© ◊ú◊§◊ô ◊©◊ù/◊û◊ô◊ô◊ú/◊ß◊ï◊®◊°‚Ä¶', layout=w.Layout(width='320px'))
dropdown = w.Dropdown(options=[], description='◊°◊ò◊ï◊ì◊†◊ò:', layout=w.Layout(width='380px'))
reload_btn = w.Button(description='◊®◊¢◊†◊ï◊ü ◊û◊î◊ß◊ï◊ë◊•', icon='refresh', button_style='', layout=w.Layout(width='140px'))
out = w.Output()

# ◊û◊ô◊ú◊ï◊ô ◊ê◊§◊©◊®◊ï◊ô◊ï◊™ ◊î◊ì◊®◊ï◊§◊ì◊ê◊ï◊ü ◊ú◊§◊ô ◊°◊ô◊†◊ï◊ü
def filtered_students(query=""):
    q = (query or "").strip().lower()
    if not q:
        return students
    res = []
    for s in students:
        hay = " ".join([
            full_name(s),
            s.get("email",""),
            " ".join(s.get("courses", []))
        ]).lower()
        if q in hay:
            res.append(s)
    return res

def build_options(items):
    opts = []
    for s in items:
        label = f"{full_name(s)}  ‚Äî  {s.get('email','')}"
        opts.append((label, s))
    return opts or [("◊ú◊ê ◊†◊û◊¶◊ê◊ï ◊™◊ï◊¶◊ê◊ï◊™", None)]

def initials(s):
    fn = s.get("first_name"," ")[:1]
    ln = s.get("last_name"," ")[:1]
    return (fn+ln).upper()

def render_student_card(s):
    name = full_name(s)
    email = s.get("email","")
    link = s.get("link","")
    courses = s.get("courses", [])

    courses_html = "".join([f"<span class='chip'>{c}</span>" for c in courses])
    avatar_txt = initials(s)

    card = f"""
    <div class="student-card">
      <div class="header">
        <div class="avatar">{avatar_txt}</div>
        <div>
          <div class="name">{name}</div>
          <div class="email"><a href="mailto:{email}">{email}</a></div>
        </div>
      </div>
      <div class="row">
        <div class="block">
          <h4>◊ß◊ï◊®◊°◊ô◊ù ◊ë◊°◊û◊°◊ò◊®</h4>
          <div class="chips">{courses_html or '<span class="chip">‚Äî</span>'}</div>
        </div>
        <div class="block">
          <h4>◊ß◊ô◊©◊ï◊® ◊û◊¢◊†◊ô◊ô◊ü</h4>
          <div class="link">{f'<a target="_blank" href="{link}">{link}</a>' if link else '‚Äî'}</div>
          <div class="help">◊ò◊ô◊§: ◊ú◊ó◊¶◊ï ◊¢◊ú ◊î◊ß◊ô◊©◊ï◊® ◊ú◊î◊§◊™◊ó ◊ë◊ú◊©◊ï◊†◊ô◊™ ◊ó◊ì◊©◊î.</div>
        </div>
      </div>
    </div>
    """
    return card

def update_dropdown(*_):
    items = filtered_students(search_box.value)
    dropdown.options = build_options(items)
    # ◊ë◊ó◊® ◊ê◊™ ◊î◊®◊ê◊©◊ï◊ü ◊ê◊ù ◊ß◊ô◊ô◊ù
    if items:
        dropdown.value = items[0]
    else:
        dropdown.value = None
    render_output()

def render_output(*_):
    out.clear_output()
    with out:
        display(HTML(style_html))  # ◊ï◊ì◊ê◊ï ◊©◊î-CSS ◊†◊ò◊¢◊ü ◊ë◊õ◊ú ◊®◊¢◊†◊ï◊ü
        s = dropdown.value
        if s:
            display(HTML(render_student_card(s)))
        else:
            display(HTML("<div style='color:#64748b'>◊ê◊ô◊ü ◊§◊®◊ô◊ò ◊ú◊î◊¶◊í◊î</div>"))

def reload_from_file(btn):
    global students
    try:
        with open(FILE_PATH, "r", encoding="utf-8") as f:
            students = json.load(f).get("students", [])
    except Exception as e:
        students = []
        out.clear_output()
        with out:
            display(HTML(f"<div style='color:#ef4444'>◊©◊í◊ô◊ê◊î ◊ë◊ß◊®◊ô◊ê◊™ ◊î◊ß◊ï◊ë◊•: {e}</div>"))
    update_dropdown()

# 7) ◊ó◊ô◊ë◊ï◊®◊ô◊ù ◊ú◊ê◊ô◊®◊ï◊¢◊ô◊ù
search_box.observe(update_dropdown, names='value')
dropdown.observe(render_output, names='value')
reload_btn.on_click(reload_from_file)

# 8) ◊î◊¶◊í◊î
ui = w.VBox([
    w.HTML("<h3 style='font-family:system-ui;margin:4px 0 8px'>üìò ◊ò◊ï◊§◊° ◊°◊ò◊ï◊ì◊†◊ò◊ô◊ù ‚Äì Braude</h3>"),
    w.HBox([
        w.HTML("<div>üîé</div>"),
        search_box,
        dropdown,
        reload_btn
    ], layout=w.Layout(gap='10px', align_items='center')),
    out
])

# ◊ê◊™◊ó◊ï◊ú ◊®◊ê◊©◊ï◊ü
update_dropdown()
display(ui)
```
